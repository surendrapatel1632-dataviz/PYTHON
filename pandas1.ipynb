```python
#code series
import pandas as pd
x=[3,4,5,6,7]
var=pd.Series(x,index=['a','b','c','d','e'],dtype="float",name="python")
print(var)
```

```python
dic={"name":['python','c','c++','java'],"poor":[12,13,14,15],"rank":[1,2,3,4]}
var1=pd.Series(dic)
print(var1)
```

```python
p=pd.Series(12,index=[1,2,3,4,5,6])
print(p)
print(type(p))
```

```python
n1=pd.Series(12,index=[1,2,3,4,5,6])
n2=pd.Series(12,index=[1,2,3,4])
print(n1+n2)
```
