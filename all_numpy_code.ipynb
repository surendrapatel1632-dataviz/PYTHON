```python
#creating array from list
import numpy as np
arr_1d=np.array([1,2,3,4,5])
print("arr_1d",arr_1d)
arr_2d=np.array([[1,2,3],[4,5,6]])
print("arr_2d",arr_2d)
```

```python
# multiply list and array
py_list=[1,2,3]
print("python list multiplication",py_list*2)
py_array=np.array([2,3,4,5,6])
print("python array multiplication",py_array*2)
```

```python
zeros=np.zeros((3,4))
print("zeros array \n",zeros)
ones=np.ones((2,3))
print("one array \n",ones)
two=np.full((2,4),2)
print("twos array\n",two)
```
